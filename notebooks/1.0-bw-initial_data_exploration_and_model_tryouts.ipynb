### How to predict the match scores for football matches:

http://11tegen11.net/2016/02/21/predicting-league-football-using-xg-and-more/

Key is to calculate the Composite Team Rating (CTR) by

CTR consisted of two separate linear models [lm in R]: one for goals scored and one for goals conceded.

It’s input is series of variables computed over the present season. This list is the same for both the model for CTR for as the model for CTR against.

Goals for & against, shots for & against, xG for & against, passes for & against, completed passes for & against, final third passes for & against, passes in the deep zone for & against, completed passes in the deep zone for & against, passes in the very deep zone for & against, completed passes in the very deep zone for & against, distance of passes into the deep zone for & against, distance of passes into the very deep zone for & against.

Also added are a few variables mixed from the present season and the previous season. The mix depends on the number of matches played in the present season. For example, after 80% of the present season, these variables consist for 20% of last season’s value and 80% of the present season’s value.

Regressed goals for & against, Regressed points per game

All of these variables, 27 in total, enter a linear model with outcome variable ‘goals scored in the next match’. A backward elimination is applied to preserve only those variables that independently correlate with the outcome.

Presently, for goals scored these are: shots for, xG for & against, complete passes in the final third, passes in the deep zone for & against, distance of passes into the very deep zone for and against, regressed goals for & against. Sometimes the direction of the correlation seems weird (f.e. xG against has a positive relation with the outcome goals scored), but this has to do with the input variable showing a lots of correlation among themselves (multicollinearity).

For goals conceded, these are: goals against, xG for & against, passes for, completed passes in the final third against, passes in the deep zone against, completed passes in the deep zone for and against, passes in the very deep zone against, regressed goals against and regressed points per game.

Offensively, these are the most important predictors according to their ‘lmg’ value in the relaimpo package of R. I’m happy to see these are all the factors that make intuitive sense to people watching football matches.

```python
import pandas as pd

# define columns of dataframe
df_2017_columns = ['team']
for ele in teams_2017['117']['history'][0].keys():
    df_2017_columns.append(ele)
    
# define dictionary to collect all data per team
df_2017 = pd.DataFrame(columns = df_2017_columns)

# extract all data per team per all single matches and put it in a data frame
for value in teams_2017.values():
    # define dummy data frame per team
    foo = pd.DataFrame(columns = df_2017_columns)
    # extract team name
    team = value['title']
    # extract match data per game day
    for i,game_day in enumerate(value['history']):
        #print(i, team)
        foo.loc[i] = [team] + [v for v in game_day.values()]
    
    df_2017 = df_2017.append(foo, ignore_index = True)
    
# taking care of the dictionaries of ppda and ppda_allowed
def extract_ppda(col,key = 'att'):
    return col[key]

for col in ['ppda', 'ppda_allowed']:
    for key in ['att', 'def']:
        df_2017[col + '_' + key] = df_2017.ppda.apply(extract_ppda, args = (key,))

df_2017 = df_2017.drop(['ppda', 'ppda_allowed'], axis=1)
```

```python
import numpy as np
def change_time_delta_into_days(col):
    return col.days + col.seconds/3600/24.0

# function to create difference columns per team
def add_resultof_timeto_next_game_per_team(df_per_team):
    # get data per team
    foo = df_per_team.copy()
    foo.sort_values('date', ascending=True, inplace = True)
    # create the difference columns
    bar = foo[['date','scored','missed']].diff().shift(-1).fillna(value = 0).rename(columns = {'date': 'time_until_next_game','scored': 'scored_diff','missed': 'missed_diff'})
    foo = pd.concat([foo,bar], axis = 1)
    # create new columns
    foo['scored_next_game'] = foo.scored + foo.scored_diff
    foo['missed_next_game'] = foo.missed + foo.missed_diff
    foo.iloc[-1, foo.columns.get_loc('scored_next_game')] = np.NaN
    foo.iloc[-1, foo.columns.get_loc('missed_next_game')] = np.NaN
    foo['time_until_next_game'] = foo['time_until_next_game'].apply(change_time_delta_into_days)
    foo = foo.drop(['scored_diff','missed_diff'], axis=1)
    # add gameday column per team
    foo['game'] = [ele + 1 for ele in list(range(foo.shape[0]))]
    return foo

# put together the global data set
teams = df_2017.team.unique()
for i,team in enumerate(teams):
    if i == 0:
        df_2017_full = add_resultof_timeto_next_game_per_team(df_2017[df_2017.team == team])
    else:
        df_2017_full = df_2017_full.append(add_resultof_timeto_next_game_per_team(df_2017[df_2017.team == team]), ignore_index=True)
        
print(df_2017_full.shape)
print(df_2017_full.columns)
df_2017_full.head()
```

```python
from datetime import datetime

### function that takes in a df and a date and gets all last game row as test/to-predict row &
### all other data before the last game as training data set without team names

def separate_train_predit_dataset(df, date = datetime.today()):
    # check if date is datetime, if not convert it
    if isinstance(date,str):
        date = datetime.strptime(date, '%Y-%m-%d')
    elif isinstance(date, datetime):
        date = date
    else:
        print( "The input date is type {0} and not in the right format (datetime or string of '%Y-%m-%d').".format(type(date)))
        
    # date all data that is before the given date
    df = df[df.date < date]
    max_game = df.game.max()
    df_train = df[df.game < max_game]
    df_predict = df[df.game == max_game]
    
    return (df_train,df_predict)

# testing the function
df_train, df_predict = separate_train_predit_dataset(df_2017_full, date = '2018-04-23')
print(df_2017_full.shape)
print(df_train.shape)
print(df_predict.shape)

assert df_predict.shape[0] == 18
```
